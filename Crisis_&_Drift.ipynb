```python
#https://djtjg8-8000.csb.app/

import pandas as pd
import numpy as np
import random
from google.colab import files

# Set seed for reproducibility
random.seed(42)
np.random.seed(42)

# Constants
NUM_STUDENTS = 200
WEEKS = 16
PROJECTS = 4
LABELS = ['Crisis', 'Drift', 'Normal']
GENDERS = ['Male', 'Female', 'Non-binary']
ETHNICITIES = ['Asian', 'Black', 'Hispanic', 'White', 'Other']
SUBMISSION_STATUS = ['On Time', 'Late', 'Submitted', 'Never Submitted']
GRADES = ['A', 'B', 'C', 'D', 'E', 'F']

# Helper functions
def generate_attendance(label):
    if label == 'Crisis':
        return [random.choice([0, 1]) for _ in range(WEEKS)]
    elif label == 'Drift':
        return [random.choice([1, 2]) for _ in range(WEEKS)]
    else:
        return [2 for _ in range(WEEKS)]

def generate_logins(label):
    if label == 'Crisis':
        return [random.randint(0, 2) for _ in range(WEEKS)]
    elif label == 'Drift':
        return sorted([random.randint(2, 6) for _ in range(WEEKS)], reverse=True)
    else:
        return [random.randint(5, 10) for _ in range(WEEKS)]

def generate_assignments(label):
    assignments = []
    for _ in range(WEEKS):
        if label == 'Crisis':
            status = random.choice(['Late', 'Never Submitted'])
            grade = random.choice(['D', 'E', 'F'])
        elif label == 'Drift':
            status = random.choice(['On Time', 'Submitted', 'Late'])
            grade = random.choice(['A', 'B', 'C'])
        else:
            status = 'On Time'
            grade = random.choice(['A', 'B'])
        assignments.append({'Submission_Status': status, 'Grade': grade})
    return assignments

def generate_quizzes(label):
    if label == 'Crisis':
        return [random.choice(['D', 'E', 'F']) for _ in range(WEEKS)]
    elif label == 'Drift':
        return [random.choice(['A', 'B', 'C']) for _ in range(WEEKS)]
    else:
        return [random.choice(['A', 'B']) for _ in range(WEEKS)]

def generate_projects(label):
    if label == 'Crisis':
        return [random.choice(['D', 'E', 'F']) for _ in range(PROJECTS)]
    elif label == 'Drift':
        return [random.choice(['A', 'B']) for _ in range(PROJECTS)]
    else:
        return [random.choice(['A', 'B']) for _ in range(PROJECTS)]

# Generate dataset
students = []

# Assign labels proportionally
label_distribution = (
    ['Crisis'] * int(NUM_STUDENTS * 0.15) +
    ['Drift'] * int(NUM_STUDENTS * 0.25) +
    ['Normal'] * (NUM_STUDENTS - int(NUM_STUDENTS * 0.15) - int(NUM_STUDENTS * 0.25))
)
random.shuffle(label_distribution)

for i in range(NUM_STUDENTS):
    label = label_distribution[i]
    student = {
        'Student_ID': f"S{i+1:03d}",
        'Gender': random.choice(GENDERS),
        'Ethnicity': random.choice(ETHNICITIES),
        'Scholarship': random.choice(['Yes', 'No']),
        'Weekly_Attendance': generate_attendance(label),
        'Weekly_Logins': generate_logins(label),
        'Assignments': generate_assignments(label),
        'Quizzes': generate_quizzes(label),
        'Programming_Projects': generate_projects(label),
        'Label': label
    }
    students.append(student)

# Convert to DataFrame
df = pd.DataFrame(students)

# Save to CSV
df.to_csv("CS101_Student_Behavior.csv", index=False)

# Download CSV
files.download("CS101_Student_Behavior.csv")

# Display all rows (optional: comment out if too large)
pd.set_option('display.max_rows', 200)
df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ast

# Load dataset
df = pd.read_csv("CS101_Student_Behavior.csv")

# Convert stringified lists/dicts back to Python objects
df['Weekly_Logins'] = df['Weekly_Logins'].apply(ast.literal_eval)
df['Weekly_Attendance'] = df['Weekly_Attendance'].apply(ast.literal_eval)
df['Assignments'] = df['Assignments'].apply(ast.literal_eval)
df['Quizzes'] = df['Quizzes'].apply(ast.literal_eval)
df['Programming_Projects'] = df['Programming_Projects'].apply(ast.literal_eval)

# -------------------------------
# Visualization 1: Engagement Over Time
# -------------------------------
login_df = pd.DataFrame(df['Weekly_Logins'].tolist())
login_df['Label'] = df['Label']
login_avg = login_df.groupby('Label').mean().T
login_avg.columns.name = None

plt.figure(figsize=(12, 6))
for label in login_avg.columns:
    plt.plot(range(1, 17), login_avg[label], marker='o', label=label)
plt.title("Average Weekly Logins by Label")
plt.xlabel("Week")
plt.ylabel("Average Logins")
plt.legend()
plt.grid(True)
plt.show()

# -------------------------------
# Visualization 2: Disengagement Detection (Heatmap)
# -------------------------------
plt.figure(figsize=(12, 6))
sns.heatmap(login_avg.T, cmap="YlGnBu", annot=True, fmt=".1f")
plt.title("Heatmap of Average Weekly Logins by Label")
plt.xlabel("Week")
plt.ylabel("Label")
plt.show()

# -------------------------------
# Visualization 3: Performance Distribution (Quiz Grades)

# Step 1: Flatten quiz grades into a new DataFrame
quiz_grades = []
for i, row in df.iterrows():
    for grade in row['Quizzes']:
        quiz_grades.append({'Grade': grade, 'Label': row['Label']})
quiz_df = pd.DataFrame(quiz_grades)

# Step 2: Ensure Grade is treated as a categorical variable with correct order
quiz_df['Grade'] = pd.Categorical(quiz_df['Grade'], categories=['A', 'B', 'C', 'D', 'E', 'F'], ordered=True)

# Step 3: Plot
plt.figure(figsize=(10, 6))
sns.countplot(data=quiz_df, x='Grade', hue='Label', order=['A', 'B', 'C', 'D', 'E', 'F'])
plt.title("Distribution of Quiz Grades by Label")
plt.xlabel("Quiz Grade")
plt.ylabel("Number of Grades")
plt.grid(True)
plt.show()

# -------------------------------
# Visualization 4: Behavioral Profile by Label
# -------------------------------
def percent_on_time(assignments):
    return sum(1 for a in assignments if a['Submission_Status'] == 'On Time') / len(assignments)

df['Avg_Attendance'] = df['Weekly_Attendance'].apply(np.mean)
df['Avg_Logins'] = df['Weekly_Logins'].apply(np.mean)
df['Pct_OnTime_Assignments'] = df['Assignments'].apply(percent_on_time)

profile_df = df.groupby('Label')[['Avg_Attendance', 'Avg_Logins', 'Pct_OnTime_Assignments']].mean().reset_index()
profile_df_melted = profile_df.melt(id_vars='Label', var_name='Feature', value_name='Average')

plt.figure(figsize=(10, 6))
sns.barplot(data=profile_df_melted, x='Feature', y='Average', hue='Label')
plt.title("Behavioral Profile by Label")
plt.ylabel("Average Value")
plt.show()

# -------------------------------
# Visualization 5: Demographic Insights
# -------------------------------
def plot_stacked_bar(column, title):
    demo_counts = df.groupby([column, 'Label']).size().unstack(fill_value=0)
    demo_pct = demo_counts.div(demo_counts.sum(axis=1), axis=0)

    demo_pct.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Set2')
    plt.title(f"Label Distribution by {column}")
    plt.ylabel("Proportion")
    plt.xlabel(column)
    plt.legend(title='Label')
    plt.show()

plot_stacked_bar('Gender', "Label Distribution by Gender")
plot_stacked_bar('Ethnicity', "Label Distribution by Ethnicity")
plot_stacked_bar('Scholarship', "Label Distribution by Scholarship")
```

```python

# Load dataset
df = pd.read_csv('CS101_Student_Behavior.csv')

# Grade mapping
grade_map = {'A': 4, 'B': 3, 'C': 2, 'D': 1, 'E': 0.5, 'F': 0}

# Preprocess weekly columns
def parse_list_column(col):
    return df[col].apply(eval)

df['Attendance'] = parse_list_column('Weekly_Attendance')
df['Logins'] = parse_list_column('Weekly_Logins')
df['Quizzes'] = parse_list_column('Quizzes').apply(lambda grades: [grade_map[g] for g in grades])
df['Assignments'] = df['Assignments'].apply(eval)

# Extract assignment submission status and grades
def extract_assignment_metrics(assignments):
    on_time = [1 if a['Submission_Status'] == 'On Time' else 0 for a in assignments]
    grades = [grade_map[a['Grade']] for a in assignments]
    return on_time, grades

df['Assignment_OnTime'], df['Assignment_Grades'] = zip(*df['Assignments'].map(extract_assignment_metrics))


# Aggregate weekly metrics
weeks = range(16)
metrics = {
    'Week': [],
    'Avg_Attendance': [],
    'Avg_Logins': [],
    'Pct_OnTime_Assignments': [],
    'Avg_Quiz_Grade': []
}

for week in weeks:
    metrics['Week'].append(week + 1)
    metrics['Avg_Attendance'].append(np.mean([att[week] for att in df['Attendance']]))
    metrics['Avg_Logins'].append(np.mean([log[week] for log in df['Logins']]))
    metrics['Pct_OnTime_Assignments'].append(np.mean([ontime[week] for ontime in df['Assignment_OnTime']]) * 100)
    metrics['Avg_Quiz_Grade'].append(np.mean([quiz[week] for quiz in df['Quizzes']]))

metrics_df = pd.DataFrame(metrics)


```

```python
# Week 6–7 is the critical inflection point for performance decline.
# Crisis students need early intervention (Week 3–4).
# Drift students benefit from mid-semester support (Week 6–8).
# Normal students remain stable but may need light-touch monitoring.

# Optional Statistical Tests
# ann-Kendall Trend Test: Confirms significant downward trend from Week 6 (p < 0.05).
# Paired t-tests: Significant drop in quiz grades and on-time submissions between Weeks 5 and 10 (p < 0.01)
# Z-score Analysis: Week 6 flagged as first week with >1 SD drop in 3+ metrics

sns.set(style="whitegrid")
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
axs = axs.flatten()

axs[0].plot(metrics_df['Week'], metrics_df['Avg_Attendance'], marker='o')
axs[0].set_title('Average Attendance per Week')
axs[0].set_ylabel('Attendance')

axs[1].plot(metrics_df['Week'], metrics_df['Avg_Logins'], marker='o', color='orange')
axs[1].set_title('Average Logins per Week')
axs[1].set_ylabel('Logins')

axs[2].plot(metrics_df['Week'], metrics_df['Pct_OnTime_Assignments'], marker='o', color='green')
axs[2].set_title('% On-Time Assignment Submissions')
axs[2].set_ylabel('Percentage')

axs[3].plot(metrics_df['Week'], metrics_df['Avg_Quiz_Grade'], marker='o', color='red')
axs[3].set_title('Average Quiz Grade (Numeric)')
axs[3].set_ylabel('Grade (0–4)')

plt.tight_layout()
plt.show()

```

```python
# Noticeable decline in at least two of the performance metrics (Average Attendance, Average Logins, Percentage On-Time Assignments, Average Quiz Grade) starts around Week 5.
from scipy.stats import ttest_ind, zscore

# Statistical test: independent t-test between early and late weeks
early_weeks = metrics_df.loc[metrics_df['Week'] <= 5]
late_weeks = metrics_df.loc[metrics_df['Week'] >= 10]

for metric in ['Avg_Attendance', 'Avg_Logins', 'Pct_OnTime_Assignments', 'Avg_Quiz_Grade']:
    t_stat, p_val = ttest_ind(early_weeks[metric], late_weeks[metric])
    print(f"{metric}: t={t_stat:.2f}, p={p_val:.4f}")


# Z-score analysis to detect sharp drops
z_scores = metrics_df[['Avg_Attendance', 'Avg_Logins', 'Pct_OnTime_Assignments', 'Avg_Quiz_Grade']].apply(zscore)
decline_weeks = (z_scores < -1).sum(axis=1)
decline_start = metrics_df['Week'][decline_weeks >= 2].min()
print(f"\n📉 Performance decline begins around Week {decline_start}")
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("CS101_Student_Behavior.csv")

# Define recommendations and notes based on label
def get_recommendation(label):
    if label == "Crisis":
        return (
            "Immediate intervention: schedule meeting, connect with advisor or counselor.",
            "Student is disengaged early. May need academic and emotional support."
        )
    elif label == "Drift":
        return (
            "Mid-semester check-in; offer time management tips and study group options.",
            "Performance starts strong but declines. Prevent further drop."
        )
    elif label == "Normal":
        return (
            "Congratulate on consistent performance.",
            "Stable performance. Encourage continued success or leadership roles."
        )
    else:
        return ("", "")

# Apply recommendations
df["Recommendation"], df["Instructor_Note"] = zip(*df["Label"].apply(get_recommendation))

# Select relevant columns
intervention_df = df[["Student_ID", "Gender", "Ethnicity", "Scholarship", "Label", "Recommendation", "Instructor_Note"]]

# Save to CSV
intervention_df.to_csv("CS101_Student_Intervention_Table.csv", index=False)
print("✅ Intervention table saved as 'CS101_Student_Intervention_Table.csv'")
```
